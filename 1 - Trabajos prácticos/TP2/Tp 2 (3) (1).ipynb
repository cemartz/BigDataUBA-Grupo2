```python
import pandas as pd
import numpy as np



base_1t2004 = pd.read_stata(r"C:\Users\luque\Desktop\t104_dta\usu_individual_T104.dta")
base_1t2024 = pd.read_excel(r"C:\Users\luque\Desktop\EPH_usu_1_Trim_2024_xls (1)\EPH_usu_1er_Trim_2024_xlsx\usu_individual_T124.xlsx")


base_1t2004.columns = base_1t2004.columns.str.upper()
base_1t2024.columns = base_1t2024.columns.str.upper()


base_1t2004["A√ëO"] = 2004
base_1t2024["A√ëO"] = 2024


base_1t2004_gba = base_1t2004[base_1t2004["REGION"] == 1]
base_1t2024_gba = base_1t2024[base_1t2024["REGION"] == 1]


variables_seleccionadas = [
    "CODUSU", "NRO_HOGAR", "COMPONENTE", "CH04", "CH06", "ESTADO", "CAT_OCUP",
    "PP3E_TOT", "P47T", "NIVEL_ED", "AGLOMERADO", "REGION", "PONDERA", "TRIMESTRE", "A√ëO"
]


base_unificada = pd.concat([
    base_1t2004_gba[variables_seleccionadas],
    base_1t2024_gba[variables_seleccionadas]
], ignore_index=True)

print("‚úÖ Base unificada creada con √©xito.")
print("üóìÔ∏è  A√±os unificados entre 2004 y 2024")
print("üìä Distribuci√≥n por a√±o:")
print(base_unificada["A√ëO"].value_counts())
print(f"üî¢ Dimensiones finales de la base antes de limpieza: {base_unificada.shape}")

print("\nüìå Tipos de datos por a√±o (antes de limpieza):")
print("\nüóìÔ∏è 2004:")
print(base_1t2004_gba[variables_seleccionadas].dtypes)
print("\nüóìÔ∏è 2024:")
print(base_1t2024_gba[variables_seleccionadas].dtypes)

na_2004 = base_1t2004_gba[variables_seleccionadas].isna().sum().reset_index()
na_2024 = base_1t2024_gba[variables_seleccionadas].isna().sum().reset_index()

na_2004.columns = ["Variable", "Faltantes_2004"]
na_2024.columns = ["Variable", "Faltantes_2024"]

na_summary = pd.merge(na_2004, na_2024, on="Variable", how='outer').fillna(0)

print("\nüìâ Tabla de valores faltantes por a√±o (antes de limpieza):")
print(na_summary.sort_values(by=["Faltantes_2004", "Faltantes_2024"], ascending=False))


print("\nüõ†Ô∏è Iniciando la correcci√≥n de valores sin sentido...")


base_unificada_limpia = base_unificada.copy()


print("\nüîÑ Tratando ingresos negativos (P47T)...")
print(f"   - Registros con ingresos negativos antes: {len(base_unificada_limpia[base_unificada_limpia['P47T'] < 0])}")
base_unificada_limpia['P47T'] = np.where(base_unificada_limpia['P47T'] < 0, np.nan, base_unificada_limpia['P47T'])
print(f"   - Registros con ingresos negativos despu√©s: {len(base_unificada_limpia[base_unificada_limpia['P47T'] < 0])}")


print("\nüîÑ Tratando horas trabajadas negativas (PP3E_TOT)...")
print(f"   - Registros con horas trabajadas negativas antes: {len(base_unificada_limpia[base_unificada_limpia['PP3E_TOT'] < 0])}")
base_unificada_limpia['PP3E_TOT'] = np.where(base_unificada_limpia['PP3E_TOT'] < 0, np.nan, base_unificada_limpia['PP3E_TOT'])
print(f"   - Registros con horas trabajadas negativas despu√©s: {len(base_unificada_limpia[base_unificada_limpia['PP3E_TOT'] < 0])}")


print("\nüîÑ Tratando edades fuera de rango (CH06)...")
print(f"   - Registros con edad < 0 antes: {len(base_unificada_limpia[base_unificada_limpia['CH06'] < 0])}")
print(f"   - Registros con edad > 100 antes: {len(base_unificada_limpia[base_unificada_limpia['CH06'] > 100])}")
base_unificada_limpia['CH06'] = np.where((base_unificada_limpia['CH06'] < 0) | (base_unificada_limpia['CH06'] > 100), np.nan, base_unificada_limpia['CH06'])
print(f"   - Registros con edad < 0 despu√©s: {len(base_unificada_limpia[base_unificada_limpia['CH06'] < 0])}")
print(f"   - Registros con edad > 100 despu√©s: {len(base_unificada_limpia[base_unificada_limpia['CH06'] > 100])}")


variables_categoricas = ["ESTADO", "CAT_OCUP", "NIVEL_ED"]
for var in variables_categoricas:
    print(f"\nüîÑ Tratando c√≥digos de no respuesta en {var}...")
    print(f"   - Registros con valor -1 en {var} antes: {len(base_unificada_limpia[base_unificada_limpia[var] == -1])}")
    print(f"   - Registros con valor -2 en {var} antes: {len(base_unificada_limpia[base_unificada_limpia[var] == -2])}")
    base_unificada_limpia[var] = np.where((base_unificada_limpia[var] == -1) | (base_unificada_limpia[var] == -2), np.nan, base_unificada_limpia[var])
    print(f"   - Registros con valor -1 en {var} despu√©s: {len(base_unificada_limpia[base_unificada_limpia[var] == -1])}")
    print(f"   - Registros con valor -2 en {var} despu√©s: {len(base_unificada_limpia[base_unificada_limpia[var] == -2])}")


print("\nüóëÔ∏è Eliminando registros con valores faltantes en las variables de inter√©s...")
base_unificada_limpia = base_unificada_limpia.dropna(subset=variables_seleccionadas)
print(f"   - Dimensiones de la base despu√©s de eliminar faltantes: {base_unificada_limpia.shape}")

print("\n‚úÖ Proceso de limpieza de valores sin sentido completado.")

```
